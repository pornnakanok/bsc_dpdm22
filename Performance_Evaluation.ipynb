# Load data

```python
from sklearn import datasets
iris_data = datasets.load_iris()
```

```python
X = iris_data['data']
y = iris_data['target']
```

# Cross validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

#Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

#Train and Evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

#Print the avarage accuracy and standard deviation scross all folds
print('Accuracy: %0.2f (+/- %0.2f)' % (scores.mean(), scores.std() * 2))
```

## Leave one out cross validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import LeaveOneOut
```

```python
# Create a LeaveOneOut cross-validation iterator
loo = LeaveOneOut()
```

```python
model = DecisionTreeClassifier()

#Define the LeaveOneOut cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

#Train and Evaluate the model using LeaveOneOut cross-validation
scores = cross_val_score(model, X, y, cv=loo)

#Print the avarage accuracy and standard deviation scross all folds
print('Accuracy: %0.2f (+/- %0.2f)' % (scores.mean(), scores.std() * 2))
```
